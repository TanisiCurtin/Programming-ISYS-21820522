Weather Data Tracking and Analysis Tool

Background - You are a junior data analyst at ClimateWatch, an organization focused on monitoring and analyzing climate patterns to support research and inform the public about climate change. Your team is dedicated to collecting and analyzing weather data from various locations to identify trends, anomalies, and potential climate impacts. You have been tasked with developing a Python application that allows users (researchers, students, and interested public members) to input and track weather data on a monthly basis for different locations

Objectives: Create a Python application using only the Python standard libraries and Matplotlib, that achieves the following functionalities:

        Data Input and Storage: Prompt the user to input monthly weather data for multiple locations. The data should include at least the location name, month, average temperature, total precipitation, and average wind speed.
        Data Analysis and Visualization: Allow the user to select from at least three analysis options (e.g., temperature trends over the year, comparison of precipitation across locations, wind speed variation through months) and visualize this data using graphs generated by Matplotlib.


Data to be entered by the user : Location Name, Month, Average Temperature, Total Precipitation and Average Speed.
Data Shown by the program: Temperature Trends for the Year, Comparison of Precipation accross Locations and Wind Speed Variation through the months using Graph MatPlotlib

```python
# Importing important libraries
import matplotlib.pyplot as plt
import calendar

#Functions to enter weather data
def input_weather_data():
  locs_data =[]
  months_list = list(calendar.month_name)[1:] #for choosing months

  while True:
    loc = input("Enter the location name or 'done'to finish:")
    if loc.lower() =='done':
      break

    print("\nSelect the month:")
    for i, month in enumerate(months_list, start=1):
      print(f"{i}.{month}")

    while True:
      try:
        month_choice=int(input("Enter the number corresponding to the month: "))
        if 1 <= month_choice <= len(months_list):
          month = months_list[month_choice - 1]
          break
        else:
          print("Invalid choice. Please enter a number within the range.")
      except ValueError:
        print("Invalid input. Please enter a number.")
    avg_temp = float(input("Enter the average temperature: "))
    total_prec = float(input("Enter the total precipitation: "))
    avg_wind_speed = float(input("Enter the average wind speed:"))

    loc_data = {
        "Location":loc,
        "Month":month,
        "Avg Temperature": avg_temp,
        "Total Precipitation": total_prec,
        "Avg Wind Speed":avg_wind_speed
    }
    locs_data.append(loc_data)
  return locs_data

#Function to analyze and visualize temperature trends over the year
def temp_trends(locs_data):
  months = [data["Month"]for data in locs_data]
  temp =[data["Avg Temperature"]for data in locs_data ]

  plt.plot(months, temp, marker='o')
  plt.title('Temperature Trends Over the Year')
  plt.xlabel('Month')
  plt.ylabel('Average Temperature(C)')
  plt.xticks(rotation=45)
  plt.grid(True)
  plt.show()

#Function to analyze and visualize comparison of precipitation across locations
def prec_comparison(locs_data):
  locs = [data["Location"]for data in locs_data]
  prec =[data["Total Precipitation"]for data in locs_data ]

  plt.plot(locs,prec)
  plt.title('Comparison of Precipitation across Locations')
  plt.xlabel('locs')
  plt.ylabel('Total Precipitation(in mm)')
  plt.xticks(rotation=45)
  plt.grid(True)
  plt.show()

#Function to analyze and visualize wind speed variation through the months
def wind_speed_variation(locs_data):
  months = [data["Month"]for data in locs_data]
  wind_speeds =[data["Avg Wind Speed"]for data in locs_data ]

  plt.plot(months,wind_speeds, marker='o')
  plt.title('Wind Speed Variation thoughout different locations')
  plt.xlabel('Month')
  plt.ylabel('Average Wind Speed(m/s)')
  plt.xticks(rotation=45)
  plt.grid(True)
  plt.show()

#Gathered the data till here now
#Main Function to run the analysis
def main():
  #input weather dta
  locations_data = input_weather_data()

  #Analysis and Visualization
  analysis_options={
      1: temp_trends,
      2: prec_comparison,
      3: wind_speed_variation
  }

  while True:
    print("\nChoose an analysis option:")
    print("1. Temperature Trends over the year")
    print("2. Comparison of precipitation across Locations")
    print("3.Wind Speed Variation through different Locations")
    print("4. Exit")

    choice= input("Enter Your Choice:")
    if choice == '4':
      break

    try:
      choice= int(choice)
      if choice in analysis_options:
        analysis_options[choice](locations_data)
      else:
        print("Invalid Choice. Please enter a Valid Option.")
    except ValueError:
      print("Invalid Input.Try to enter a number from 1-4")

if __name__=="__main__":
  main()
```
